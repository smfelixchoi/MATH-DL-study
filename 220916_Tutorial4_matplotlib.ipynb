## 함수 그래프 그리기

```python
x = np.arange(0, 6.28, 0.01)
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
z = np.cos(x)
```

```python
plt.plot(x, y, label='sine')
plt.plot(x, z, linestyle='--', label='cosine')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Sine and Cosine curves')
plt.legend()
plt.show()
```

## Random Walk 그려보기

- numpy array의 `cumsum()` 메소드를 이용하여 그려봅니다.

```python
a = np.arange(10)
```

```python
# cumulative sum
a.cumsum()
```

```python
x = np.arange(10000)
z = np.random.randn(10000)
```

```python
z_sum = z.cumsum()
```

```python
plt.plot(x,z_sum)
plt.show()
```

```python
x = np.arange(1000)

for _ in range(10):
    temp = np.random.randn(1000)
    temp = temp.cumsum()
    plt.plot(x, temp)
    
plt.show()
```

## subplot 이용해서 따로 그리기

```python
fig, axes = plt.subplots(nrows=2, ncols=3)

for i in range(2):
    for j in range(3):
        axes[i,j].plot(np.random.rand(100))
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3)

for i in range(2):
    for j in range(3):
        axes[i,j].hist(i*5+(j+1)*np.random.randn(5000), bins=500)
```
